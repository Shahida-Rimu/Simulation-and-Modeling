```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
# Sample scalar and array inputs
scalar = 3.7
arr = np.array([12, 22, 13, 4, 25, 33])

# Trigonometric functions
sin_values = np.sin(arr)
cos_values = np.cos(arr)
tan_values = np.tan(arr)
asin_values = np.arcsin(np.clip(arr / np.max(arr), -1, 1))  # Normalized and Clipping for valid range [-1, 1]
acos_values = np.arccos(np.clip(arr / np.max(arr), -1, 1))
atan_values = np.arctan(arr)

# Exponential and logarithm
exp_values = np.exp(arr)
log_values = np.log(arr)  # Natural logarithm

# Absolute value and square root
abs_values = np.abs(arr)
sqrt_values = np.sqrt(arr)

# Remainder when divided by a scalar (e.g., 2)
rem_values = np.remainder(arr, 2)

# Rounding functions
round_values = np.round(arr)
floor_values = np.floor(arr)
ceil_values = np.ceil(arr)

# Display results
print("sin:", sin_values)
print("cos:", cos_values)
print("tan:", tan_values)
print("asin:", asin_values)
print("acos:", acos_values)
print("atan:", atan_values)
print("exp:", exp_values)
print("log:", log_values)
print("abs:", abs_values)
print("sqrt:", sqrt_values)
print("rem:", rem_values)
print("round:", round_values)
print("floor:", floor_values)
print("ceil:", ceil_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define a 2Ã—3 matrix with float values
mat = np.array([[2.1, 4.7, 1.5],
                [3.3, 0.9, 5.8]])

# Flattened version for operations like max, min, etc.
flat_mat = mat.flatten()

# 1. Max & Min values with indices
max_value = np.max(flat_mat)
max_index = np.argmax(flat_mat)
min_value = np.min(flat_mat)
min_index = np.argmin(flat_mat)

# Convert flat index to 2D index
max_index_2d = np.unravel_index(max_index, mat.shape)
min_index_2d = np.unravel_index(min_index, mat.shape)

# 2. Length of the "vectorized" matrix (total elements)
vec_length = flat_mat.size

# 3. Sorted version
sorted_entire_desc = np.sort(flat_mat)[::-1]  # Entire matrix sorted in descending order
sorted_rowwise_desc = -np.sort(-mat, axis=1)  # Row-wise descending sort

# 4. Sum & Product
sum_values = np.sum(flat_mat)
prod_values = np.prod(flat_mat)

# 5. Median, Mean, Standard Deviation
median_value = np.median(flat_mat)
mean_value = np.mean(flat_mat)
std_dev = np.std(flat_mat)

# Display Results
print("Original Matrix:\n", mat)
print("Max value:", max_value, "at index", max_index_2d)
print("Min value:", min_value, "at index", min_index_2d)
print("Length of matrix (flattened):", vec_length)
print("Sorted (entire matrix, descending):", sorted_entire_desc)
print("Sorted (row-wise, descending):\n", sorted_rowwise_desc)
print("Sum of elements:", sum_values)
print("Product of elements:", prod_values)
print("Median value:", median_value)
print("Mean value:", mean_value)
print("Standard deviation:", std_dev)

```

```python

# Replace static matrix with random numbers
random_matrix = np.random.uniform(0, 10, size=(3, 2)) # Generate a 3x2 matrix with random floats between 0 and 10

# Perform operations on the random matrix
sin_values_rand = np.sin(random_matrix)
cos_values_rand = np.cos(random_matrix)
tan_values_rand = np.tan(random_matrix)
asin_values_rand = np.arcsin(np.clip(random_matrix / np.max(random_matrix), -1, 1))
acos_values_rand = np.arccos(np.clip(random_matrix / np.max(random_matrix), -1, 1))
atan_values_rand = np.arctan(random_matrix)
exp_values_rand = np.exp(random_matrix)
log_values_rand = np.log(random_matrix)
abs_values_rand = np.abs(random_matrix)
sqrt_values_rand = np.sqrt(random_matrix)
rem_values_rand = np.remainder(random_matrix, 2)
round_values_rand = np.round(random_matrix)
floor_values_rand = np.floor(random_matrix)
ceil_values_rand = np.ceil(random_matrix)

# Display results for the random matrix
print("\nRandom Matrix:\n", random_matrix)
print("\nsin (random):\n", sin_values_rand)
print("\ncos (random):\n", cos_values_rand)
print("\ntan (random):\n", tan_values_rand)
print("\nasine (random):\n", asin_values_rand)
print("\nacosine (random):\n", acos_values_rand)
print("\natangent (random):\n", atan_values_rand)
print("\nexp (random):\n", exp_values_rand)
print("\nlog (natural, random):\n", log_values_rand)
print("\nabs (random):\n", abs_values_rand)
print("\nsqrt (random):\n", sqrt_values_rand)
print("\nrem (remainder when divided by 2, random):\n", rem_values_rand)
print("\nround (random):\n", round_values_rand)
print("\nfloor (random):\n", floor_values_rand)
print("\nceil (random):\n", ceil_values_rand)


# Visualize the random matrix using a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(random_matrix, annot=True, cmap="viridis", fmt=".2f", linewidths=.5)
plt.title("Heatmap of Random Matrix Values")
plt.xlabel("Column Index")
plt.ylabel("Row Index")
plt.show()
```
